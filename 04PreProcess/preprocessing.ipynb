```python
import pandas as pd
```

```python
df = pd.read_csv('./data/RptManagerList.csv')
```

```python
people = df['姓名']
```

```python
experience = df['经历']
```

```python
experience.to_csv('./data/experiences.csv', index=False)
```

```python
people.to_csv('./data/people.csv', index=False)
```

```python
import sys
```

```python
max_length = 0;
max_exp_idx = 0;
exp = experience.tolist();
```

```python
for idx, i in enumerate(exp):
    length = len(i)
    if length > max_length:
        max_length = length
        max_exp_idx = idx
```

```python
max_length
```

```python
max_words = exp[max_exp_idx]
```

```python
max_words
```

```python
max_size = sys.getsizeof(max_words)
```

```python
print(max_size)
```
