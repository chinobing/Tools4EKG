```python
import pandas as pd
```

```python
df = pd.read_csv('./data/CompanySurvey/jbzl.csv')
```

```python
df
```

```python
check = df['律师事务所']
```

```python
check
```

```python
check.unique()
```

```python
lawyer = list(check.unique())
```

```python
lawoffice = []
for item in lawyer:
    words = item.split(',')
    for word in words:
        lawoffice.append(word)
lawoffices = set(lawoffice)
print(lawoffices)
```

```python
thefile = open('./data/CompanySurvey/lawoffice.txt', 'w')
for item in lawoffices:
    thefile.write("%s\n" % item)
thefile.close()
```

```python
check = df['会计师事务所']
```

```python
check
```

```python
check.unique()
```

```python
account = list(check.unique())
```

```python
accountoffice = []
for item in account:
    words = item.split(',')
    for word in words:
        accountoffice.append(word)
accountoffices = sorted(set(accountoffice))
print(accountoffices)
```

```python
thefile = open('./data/CompanySurvey/accountoffice.txt', 'w')
for item in accountoffices:
    thefile.write("%s\n" % item)
thefile.close()
```
